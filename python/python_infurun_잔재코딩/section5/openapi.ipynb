```python

import requests
import pprint
import openpyxl


client_id = 'Yb_YbPdPar41rThapbH9'

clinet_secret = 'VvHFHHQn7W'

# url의 ?는 쿼리스트링의 시작을 의미한다. 
# $는 여러개의 파라미터를 전달할 수 있다. 

"""
쿼리 스트링 설명 

https://openapi.naver.com/v1/search/shop.json?query=macbookpro&display=100&start=1

위의 url의 의미는 맥북 프로를 검색을 하는데 display=100은 100개의 데이터를 출력하고, 그 시작은 1번째부터 출력한다.

만약, start=101로 되어 있으면, 101번째부터 100개를 출력하므로, 101 ~ 201번째까지 출력 

"""

start_num,num = 1 , 0

excel_file = openpyxl.Workbook()

excel_sheet = excel_file.active

excel_sheet.column_dimensions['B'].width = 100 
excel_sheet.column_dimensions['C'].width = 100 
excel_sheet.append(['랭킹','제목','링크'])

for index_num in range(10):

    naver_open_api = f'https://openapi.naver.com/v1/search/shop.json?query=macbookpro&display=100&start={start_num + (index_num * 100) }'
    
    header_params = {"X-Naver-Client-Id" : client_id, "X-Naver-Client-Secret" : clinet_secret}
    
    # http 규약에 따라, 주소와 헤더의 부분을 넣는다. 
    res = requests.get(naver_open_api,headers=header_params)
    # 응답 본문이 json 형식일 경우, 이를 파이썬 딕셔너리나 리스트로 자동으로 변환하여 반환한다. 
    pprint.pprint(res.json())

    if res.status_code == 200:
        data = res.json()
        for item in data['items']:
            num += 1
            print(f"""{num,item['title']}
            
            {item['link']}""")
            excel_sheet.append(num,item['title'],item['link'])
    else : 
        print("Error Code : ", res.status_code)
excel_file.save('shopping_with_macbook_pro.xlsx')
excel_file.close()
# content는 응답 본문을 바이트 형태로 반환한다. 이 데이터를 이진 데이터이거나 인코딩된 텍스트일 수 있다. ㄷ
# print(res.content)
```
