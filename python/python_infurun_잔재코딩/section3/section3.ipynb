## 크롤링 ( Crawling ) 

### 크롤링에 필요한 라이브러리 

* requests
 웹 페이지 가져오기 라이브러리
* bs4
  웹 페이지 분석 ( 크롤링 ) 라이브러리

```python
import requests
from bs4 import BeautifulSoup # bs4 라이브러리에서 BeautifulSoup라는 클래스를 사용하겠다. 

# import web browser 

res = requests.get("https://www.naver.com")
res.content
```

```python
# web page parsing 

soup = BeautifulSoup(res.content, 'html.parser') 
# soup.find() 함수로 원하는 부분을 지정할 수 있다. 
# 변수.get_text() 함수로 추출한 부분을 가져올 수 있다. 

mydata = soup.find('title') # HTML 코드로 표시된다. 
print(mydata.get_text()) # 텍스트만 추출 
```

```python
from bs4 import BeautifulSoup
html = "<html> \
<body> \
<h1 id='title'>[1]크롤링이란?</h1> \
<p class='cssstyle'>웹페이지에서 필요한 데이터를 추출하는 것</p> \
<p id='body' align='center'>파이썬을 중심으로 다양한 웹크롤링 기술 발달</p> \
</body> \
</html>"
soup = BeautifulSoup(html,
"html.parser")
# 태그로 검색 방법
data = soup.find('h1')
print(data)
print(data.string)
print(data.get_text())
```
