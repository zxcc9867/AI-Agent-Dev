### 1단계: 회사명만 출력해보기 

````python
import requests
from bs4 import BeautifulSoup

res = requests.get('https://davelee-fun.github.io/blog/crawling_stock_example.html')
soup = BeautifulSoup(res.content, 'html.parser')
data = soup.select("ul.table-stock > li.row_sty")

for item in data:
    print (item.select_one("div.st_name").get_text().replace(" ", "").replace("\n", ""))
````

### 2단계: 회사명과 현재금액 한 줄에 출력해보기

````python
import requests
from bs4 import BeautifulSoup

res = requests.get("https://davelee-fun.github.io/blog/crawling_stock_example.html")
soup = BeautifulSoup(res.content, "html.parser")
data = soup.select("ul.table-stock > li.row_sty")

for item in data:
    price = item.select_one("span.price").get_text().replace(" ", "").replace("\n", "")
    print(item.select_one("div.st_name").get_text().replace(" ", "").replace("\n", "")," : ",price)
    
````

### 3단계: 회사명, 현재금액, 변동률 한 줄에 출력해보기

````python
import requests
from bs4 import BeautifulSoup

## 함수를 만들어서 패턴을 처리

def create_text_pattern(text):
    return text.get_text().replace(" ", "").replace("\n", "")


res = requests.get("https://davelee-fun.github.io/blog/crawling_stock_example.html")
soup = BeautifulSoup(res.content, "html.parser")
data = soup.select("ul.table-stock > li.row_sty")

for item in data:
    price = create_text_pattern(item.select_one("span.price"))
    name = create_text_pattern(item.select_one("div.st_name"))
    fluctuation_rate = create_text_pattern(item.select_one("div.st_rate > span.icon"))
    print(name,"주식 가격은 :",price,fluctuation_rate)
````
