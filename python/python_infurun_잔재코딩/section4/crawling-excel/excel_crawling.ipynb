```python
def excet_crawling(filename, sheetname, listdata ) :
    # create excel 
    excel_file = openpyxl.Workbook()
    # activate sheet 
    excel_sheet = excel_file.active
    # A열의 너비를 100으로 설정한다. 
    # 너비를 넓힘으로써, 텍스트가 잘리지 않고, 잘 보일 수 있도록 한다. 
    excel_sheet.column_dimensions['A'].width = 100 
    # sheet name이 있다면, 
    if sheetname != '':
        # 액셀 시트의 타이틀을 설정한다. 
        excel_sheet.title = sheetname 
    for item in listdata: 
        excel_sheet.append(item)
    # filename의 매개변수 이름으로 저장 
    excel_file.save(filename)
    excel_file.close()
    
    
```

```python

# 크롤링 연습 

"""
연습 사이트 

https://davelee-fun.github.io/
"""
import requests
from bs4 import BeautifulSoup as bp
url = 'https://davelee-fun.github.io/'
res = requests.get(url)

data = bp(res.content,'html.parser')

find_data = data.find_all('h4','card-text')

find_data = data.select('div.col-md-6.mb-30px h4.card-text')

for item in find_data:
    print(item.get_text())
```

```python
# 여러 페이지 크롤링 

import requests
from bs4 import BeautifulSoup # html과 xml 파싱할 때 사용 
import openpyxl
url = 'https://davelee-fun.github.io/'
res = requests.get(url)


# 10번째 페이지까지 출력 
for page in range(10):
    
    if page >= 1:
        url = f"https://davelee-fun.github.io/page{page}/"
    data = BeautifulSoup(res.content,'html.parser')
    data_detail = list(data.select('h4.card-text'))
    for item in data_detail:
        print(item.get_text().strip())

# 액셀 파일에 크롤링한 데이터 넣기 

# 워크북 생성 
excel_file = openpyxl.Workbook()
# 액셀 시트 활성화 
excel_sheet = excel_file.active
# 액셀 시트 크기 조정 
excel_sheet.column_dimensions['A'].width = 100
excel_sheet.column_dimensions['B'].width = 20
sheetname = '' 
sheetname = sheetname or 'default_name'
for item in data_detail:
    # print(item) # item은 태그 객체이므로, 리스트에 넣을 수 없다. ㄷ
    excel_sheet.append([item.get_text().strip()])
excel_file.save('crawling')
excel_file.close()
```
