### 정규 표현식 연습 

```python
import re
string = " (Dave) "
"""
string에서 대문자 A부터 Z , 소문자 a ~ z , 숫자0 ~ 9가 아닌 것을 공백으로 바꾸어라. 
^ 표시가 있으므로, 해당 조건의 반대를 의미한다. 

[ ]와 같은 대괄호는 대괄호 안에 포함된 문자 중 하나라도 매치되면, 그 문자가 해당 문자열에서 검색이 된다. 

예를 들어, [abc] 는 'a', 'b' , 'c' 중 하나와 일치하는 문자들을 찾는다. 
"""
re.sub('[^A-Za-z0-9]','',string)
```

```python
import re

pattern = re.compile('D.A')

"""

. 은 메타문자로, 줄바꿈 문자일 \n을 제외한 모든 문자 ( 한개 )를 의미한다. 
즉, D.A는 D+모든문자 ( 한개 )+ A를 의미한다. 
DAA , DvA ,D1A

"""

pattern.search("DAA")
```

### 반복 

- '?' 는 앞 문자가 0번또는 1번 표시되는 패턴 ( 없어도 되고, 한번 있어도 되는 패턴 )
- '*' 는 앞 문자가 0번 또는 그 이상 반복되는 패턴 
- '+' 는 앞 문자가 1번 또는 그 이상 반복되는 패턴 

```python
# 정규 표현식 연습

"""
re.compile()은 정규식 패턴 객체를 생성한다.
re.sub()는 일치하는 패턴을 다른 텍스트로 변환할 때 사용한다. 
"""

# 앞에 문자 D가 없거나, 여러번 반복되고 마지막이 A인 문자열
pattern = re.compile('D?A')

# 모두 일치 
pattern.search("A2")
pattern.search("A")
pattern.search("2DA") # 1 ~ 3번인덱스가 매치
```

```python
"""
문자열 앞에 D가 없거나 여러번 반복되고, 마지막이 A인 문자열
"""

pattern = re.compile("D*A")
pattern2 = re.compile("D?A")

pattern.search("DDDaaaA")
pattern.search("DDDA") 
pattern2.search("DDA")
```

```python
"""
앞의 문자가 한번은 나와야한다. 
"""

pattern = re.compile('D+A') # D가 한번 이상은 나와야 한다.

pattern.search("DA") 
pattern.search("D1A") # 일치안됨
pattern.search("A") # 일치 안됨 
pattern.search("DDDDDDDDDDDAAA") # 일치한다. 0 ~ 12번인덱스까지 
```

### 또다른 반복 표현 : {n} , {m:n}

- {n} : 앞 문자가 n번 반복되는 패턴 
- {m,n} : 앞 문자가 m번 반복되는 패턴부터 n번 반복되는 패턴까지 

```python
pattern = re.compile('[AD]{3}A') # A또는 D 중 하나의 문자를 의미 3번 반복되는 패턴과 마지막에 A가 있는 것 
pattern.search("ADA") # 일치 안된다.
pattern.search("ADADAD") # 일치 안된다. 
pattern.search("ADADADA") # 일치하는 패턴 첫번째를 출력한다. 

pattern2 = re.compile('AD{2}A')
print(pattern2.search("ADDA")) # 첫번째 문자로 A가 있고 D가 반복되는 패턴 
print(pattern2.search('DDA')) # 첫번째 문자로 A가 없기 때문에 일치하지 않는다. 

"""
[AD]{3}: A 또는 D 중 하나의 문자가 3번 반복되는 패턴을 의미합니다. 예를 들어, AAA, ADA, DDA, AAD 등이 포함될 수 있습니다.
마지막 A: [AD]{3} 뒤에 A가 반드시 와야 합니다.
"""

## 문자열 반복 패턴

# AD라는 문자열이 2번반복되고, 마지막에 A가 있는 것을 패턴 
pattern3 = re.compile('(AD){2}A')
pattern3.search("ADADA") # 일치 한다. 
pattern3.search("ADAD2A") # 일치 하지 않는다. 
```

### match와 search 함수에 대해서 

- match : 문자열 처음부터 정규식과 매칭되는 패턴을 찾아서 리턴 
- search : 문자열 전체를 검색해서 정규식과 매칭되는 패턴을 찾아서 리턴 

```python
"""
match와 search의 차이 
"""
pattern = re.compile('[a-z]+') # 소문자 a ~ z까지 문자가 나오는 것이 한번이상있으면 매치 

matched = pattern.match('Dave') # 매치가 안된다. 대문자D가 있기 때문에 매치가 안됨

searched = pattern.search('Dave') # 매치가 된다. D가 매치되지 않더라도, 그 다음에 매치되는 패턴인 ave가 매치가 된다.

print(matched,searched)

# None <re.Match object; span=(1, 4), match='ave'>
```

### findall 함수 

- 정규 표현식과 매칭되는 모든 문자열을 리스트 객체로 리턴한다.

```python
import re 

pattern = re.compile('[a-z]+') # 소문자 a~z까지 한번이상 나오는 패턴 
findalled = pattern.findall('Hello World!') # 매칭되는 패턴을 모두 출력 
print(findalled)

pattern2 = re.compile('[a-zA-Z]+')
findalled2 = pattern2.findall('Game of Life in Python')
print(findalled2) # 이렇게 문자열을 추출할 수 있다. 
```

```python
pattern2 = re.compile(':') # 라는 패턴을 컴파일 

splited = pattern2.split('python:kava') # :를 기분으로 문자열을 분할 

print(splited)
```
