```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter

loader = PyPDFLoader("data/2040_seoul_plan.pdf")
data_nyc = loader.load()

# 문서를 청크 단위로 분할
# chunk_size: 청크의 최대 크기
# chunk_overlap: 청크 간의 겹치는 부분 크기 -> 오버랩을 통해 앞 청크 문장을 들고와서, 문맥을 유지하여, AI가 판단할 수 있게함.
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=0)
all_splits = text_splitter.split_documents(data_nyc)

for i, split in enumerate(all_splits):
    print(f"=== Split {i+1} ===")
    print(split)
```

```python
# 위 코드를 실행해보면, 중복되는 내용 없이 페이지 단위로 청크가 끊어져 있다.
# 각 청크를 오버랩을 포함하여, 작성하기 위해, 강제로 어ㅗ버랩을 처리 

for i in range(len(all_splits) -1):
    all_splits[i].page_content += " " + all_splits[i+1].page_content[:100]  # 각 청크의 끝에 다음 청크의 처음 100자를 추가
```
